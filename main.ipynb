**Authentication of the account - Checkng if the twitter developer account credentials are correct**

```python
import tweepy
import time

# assign the values accordingly
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

# Authenticate to Twitter
# authorization of consumer key and consumer secret
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)

# set access to user's access key and access secret
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

try:
    api.verify_credentials()
    print("Authentication OK")
except:
    print("Error during authentication")
```

**Finding number of unique followers with bio consisting of word "Mom" or "dad" or "parent"**

```python
import re
pattern1 = r"[^a-z]mom[^e]"
pattern2 = r"[^a-z]dad"
pattern3= r"[^a-z]parent"
for i in bio:
    if (re.search(pattern1,i.lower()) is not None) or (re.search(pattern2,i.lower()) is not None) or (re.search(pattern3,i.lower()) is not None): 
        print(i)
        bio2.append(i)
```

**Finding number of unique followers with bio consisting of word "soft skills" or "skills" or "SEL"**

```python
import re
pattern1 = r"soft skills"
pattern2 = r"SEL"
pattern3= r"skills"
for i in bio:
    if (re.search(pattern1,i.lower()) is not None) or (re.search(pattern2,i) is not None) or (re.search(pattern3,i.lower()) is not None): 
        print(i)
        bio3.append(i)
```

**Finding number of unique followers with bio consisting both the above conditions as TRUE**

```python
for i in bio3:
    if i in bio2:
        bio4.append(i)
print(bio4)
```
